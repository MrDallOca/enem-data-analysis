```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 200
sns.set_style("whitegrid")
```

```python
conn = sqlite3.connect('microdados_enem2017/database.db')
```

```python
df1 = pd.read_sql('SELECT nu_nota_mt,tp_cor_raca,q006,sg_uf_residencia,q001,q002 FROM enem', conn).apply(pd.to_numeric)
```

```python
sns.set(font_scale=0.5)
g = sns.barplot(y='NU_NOTA_MT', x='Q002', data=df1, errwidth=0, 
                palette=['#ffc299', '#ffb380', '#ffa366', '#ff944d', '#ff8533', '#ff751a', '#ff6600', '#808080'])
g.set(xlabel ="Nível de escolaridade da mãe", ylabel ="Nota média de matemática (0-1000)")
g.set_xticklabels(["Nunca estudou", 
                   "Não completou 5º ano", 
                   "Completou 5º ano, mas não 9º", 
                   "Completou 9º ano, mas não EM", 
                   "Completou EM, mas não facul.", 
                   "Completou facul., mas não pós", 
                   "Completou pós", 
                   "Não sei"],
                  rotation=75)
```
